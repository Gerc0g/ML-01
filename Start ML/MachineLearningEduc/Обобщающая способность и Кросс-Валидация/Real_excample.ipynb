```python
#Воспользуемся линейной регрессией
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)

print('MSE равен: ', int(np.mean((y-model.predict(x))**2)))
print('RMSE равен: ', int(np.mean((y-model.predict(x))**2)**(1/2)))


#RMSE берется корень чтобы порядок MSE значения был приближен к нашим таргетам
```

```python
#Замерим качество на Кросс-Валидации

#Метод KFOLD

from sklearn.model_selection import KFold

kf = KFold(n_splits = 10, shuffle = True, random_state= 33)
#n_splits - колличеств разбиений нашей выборке
#shuffle - перемешаем данные, для избежания выбросов в одном фрагменте выборки-сплита
#random_state - добавляем момет случайности в процессе обучения моели, = 33 - это параметр инициализации генератора случайных чисел
```

```python
kf.get_n_splits() #На какое количество частей подеим нашу выборку
```

```python
kf_index_temp = kf.split(x)
#По сути мы передаем список индексов отдельно для теста отельно для тренировки и передаем так чтобы индексы нигде не повторялись
print(kf_index_temp)
```

```python
i = 0
for train_index, test_index in kf.split(x):
    i += 1
    print(f'Номер разбиения:{i} - {train_index}, {test_index}')
```

```python
#
#x.values - возвращет из pd таблицы по которой мы не можем индексироваться np.array

losses_train = []
losses_test = []


for train_index, test_index in kf.split(x):
    X_train , X_test = x.values[train_index] , x.values[test_index]
    Y_train, Y_test = y.values[train_index], y.values[test_index]
    
    #Каждую итерацию мы получаем массивы тестовой и тернировочной выборки
    #Также идентифицируем линейную регрессию, оубчем и смотрим ошибки
    model = LinearRegression()
    model.fit(X_train, Y_train)
    #Качество соответственно будем смотреть на тесте
    losses_test.append(np.mean((model.predict(X_test)-Y_test)**2)**(1/2))
    losses_train.append(np.mean((model.predict(X_train)-Y_train)**2)**(1/2))
```
